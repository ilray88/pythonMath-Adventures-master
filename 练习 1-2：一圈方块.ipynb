```python
import turtle
#Ex. 1-2:

from turtle import *

def square():
    for i in range(4):
        forward(100)
        right(90)

for i in range(60):
    square()
    right(5)
ts = turtle.getscreen()
ts.getcanvas().postscript(file=r"circle.eps")
from PIL import Image
from IPython.core.display import display
display(Image.open('circle.eps'))
turtle.exitonclick()
```
