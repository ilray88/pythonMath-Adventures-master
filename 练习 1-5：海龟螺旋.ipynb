```python
import turtle
from turtle import *
#Ex. 1-5

def square(sidelength):
    '''Draws a square of a
    given sidelength'''
    for i in range(4):
        forward(sidelength)
        right(90)

def spiral():
    length = 5
    for i in range(60):
        square(length)
        right(5)
        length += 5
square(10)
spiral()
```

```python
ts = turtle.getscreen()
ts.getcanvas().postscript(file=r"circle.eps")
from PIL import Image
from IPython.core.display import display
display(Image.open('circle.eps'))
turtle.exitonclick()
```
